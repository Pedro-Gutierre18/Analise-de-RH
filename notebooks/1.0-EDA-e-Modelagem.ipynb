```python
# 1. IMPORTAÇÃO DAS BIBLIOTECAS

# Para manipulação de dados
import pandas as pd
import numpy as np

# Para visualização de dados
import matplotlib.pyplot as plt
import seaborn as sns

# Configurações de visualização para os gráficos
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 8)

print("Bibliotecas importadas com sucesso!")


# 2. CARREGAMENTO E INSPEÇÃO INICIAL DOS DADOS

caminho_arquivo = '../data/HR-Employee-Attrition.csv'
df = pd.read_csv(caminho_arquivo)

# Exibir as 5 primeiras linhas para entender a estrutura
print("As 5 primeiras linhas do dataset:")
display(df.head())

# Exibir informações gerais do DataFrame (tipos de dados, valores não nulos)
print("\nInformações gerais do DataFrame:")
df.info()

# Exibir um resumo estatístico das colunas numéricas
print("\nResumo estatístico das colunas numéricas:")
display(df.describe())
```

```python
# 3. ANÁLISE EXPLORATÓRIA DE DADOS (EDA)

# Qual é a taxa de rotatividade geral?

# Cálculo da Taxa de Rotatividade em porcentagem
turnover_rate = df['Attrition'].value_counts(normalize=True) * 100

print("Taxa de Rotatividade (%):")
print(turnover_rate)

# Visualização da Distribuição de Attrition
plt.figure(figsize=(7, 5)) 
sns.countplot(x='Attrition', data=df, palette=['skyblue', 'salmon'])

# Adicionando título e rótulos
plt.title('Distribuição de Rotatividade de Funcionários', fontsize=16)
plt.xlabel('Rotatividade (Attrition)', fontsize=12)
plt.ylabel('Número de Funcionários', fontsize=12)

# Exibe o gráfico
plt.show()
```

```python
# Quais departamentos possuem a maior taxa de rotatividade?

# Usando crosstab para criar uma tabela de frequência
turnover_by_dept = pd.crosstab(df['Department'], df['Attrition'], normalize='index') * 100 # normalize='index' calcula a porcentagem por linha (por departamento)

# Arredondando para 2 casas decimais
turnover_by_dept = turnover_by_dept.round(2)

# Ordenando os resultados pela coluna 'Yes' para ver os piores casos primeiro
turnover_by_dept = turnover_by_dept.sort_values(by='Yes', ascending=False)

print("Taxa de Rotatividade (%) por Departamento:")
display(turnover_by_dept)

# Visualização da taxa de rotatividade por departamento
plt.figure(figsize=(10, 6))

# Gráfico de barras apenas com a porcentagem de 'Yes' (saídas)
sns.barplot(x=turnover_by_dept.index, y=turnover_by_dept['Yes'], palette='viridis')

plt.title('Taxa de Rotatividade ("Sim") por Departamento', fontsize=16)
plt.xlabel('Departamento', fontsize=12)
plt.ylabel('Taxa de Rotatividade (%)', fontsize=12)
plt.xticks(rotation=45) # Rotaciona os nomes dos departamentos para melhor leitura

plt.show()
```

```python
# Existe correlação entre tempo de casa, idade, salário e a rotatividade?

# Para analisar a correlação, precisamos de uma versão numérica da nossa variável alvo 'Attrition'.
df['Attrition_numeric'] = df['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)

# Análise com Heatmap de Correlação
correlation_vars = ['Age', 'YearsAtCompany', 'MonthlyIncome', 'Attrition_numeric']
corr_matrix = df[correlation_vars].corr()

print("Matriz de Correlação:")
display(corr_matrix)

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap de Correlação entre Variáveis Chave', fontsize=16)
plt.show()

# Análise Visual com Boxplots 
fig, axes = plt.subplots(1, 3, figsize=(20, 7))

# Gráfico 1: Idade vs. Rotatividade
sns.boxplot(ax=axes[0], x='Attrition', y='Age', data=df, palette='viridis')
axes[0].set_title('Idade vs. Rotatividade')

# Gráfico 2: Tempo de Casa vs. Rotatividade
sns.boxplot(ax=axes[1], x='Attrition', y='YearsAtCompany', data=df, palette='viridis')
axes[1].set_title('Tempo de Casa vs. Rotatividade')

# Gráfico 3: Salário Mensal vs. Rotatividade
sns.boxplot(ax=axes[2], x='Attrition', y='MonthlyIncome', data=df, palette='viridis')
axes[2].set_title('Salário Mensal vs. Rotatividade')

# Ajusta o layout para evitar sobreposição de títulos
plt.tight_layout()
plt.show()
```

```python
# Quais outros fatores influenciam na saída de funcionários?

fig, axes = plt.subplots(1, 2, figsize=(18, 7))

# Gráfico 1: Rotatividade por Horas Extras (OverTime)
sns.countplot(ax=axes[0], x='OverTime', hue='Attrition', data=df, palette='Set1')
axes[0].set_title('Rotatividade por Horas Extras', fontsize=14)
axes[0].set_xlabel('Faz Horas Extras?', fontsize=12)
axes[0].set_ylabel('Número de Funcionários', fontsize=12)

# Gráfico 2: Rotatividade por Satisfação no Trabalho (JobSatisfaction) 
sns.countplot(ax=axes[1], x='JobSatisfaction', hue='Attrition', data=df, palette='Set2')
axes[1].set_title('Rotatividade por Satisfação no Trabalho', fontsize=14)
axes[1].set_xlabel('Nível de Satisfação (1=Baixo, 4=Alto)', fontsize=12)
axes[1].set_ylabel('Número de Funcionários', fontsize=12)

plt.tight_layout()
plt.show()

# Para confirmar a intuição do gráfico de OverTime, vamos ver os percentuais:
print("Percentual de saída para quem faz e não faz hora extra:")
display(pd.crosstab(df['OverTime'], df['Attrition'], normalize='index') * 100)
```

```python
# 4. CONSTRUÇÃO DO MODELO DE MACHINE LEARNING
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Pré-processamento dos Dados para o Modelo

# Remover colunas que não mais úteis
df_model = df.drop(['EmployeeCount', 'StandardHours', 'Over18', 'EmployeeNumber', 'Attrition'], axis=1)

# Transformando as colunas de texto em colunas numéricas (0s e 1s)
df_dummies = pd.get_dummies(df_model, drop_first=True)

# Separar as features (X) da variável alvo (y)
X = df_dummies.drop('Attrition_numeric', axis=1)
y = df_dummies['Attrition_numeric']

# Divisão dos Dados em Treino e Teste

# Vamos dividir nossos dados: 70% para o modelo aprender (treino) e 30% para testar o que ele aprendeu (teste)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y) # stratify=y garante que a proporção de 'Yes' e 'No' seja a mesma em treino и teste.

# Treinamento do Modelo Random Forest
model = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced') # class_weight='balanced' ajuda o modelo a lidar com dados desbalanceados, dando mais importância à classe minoritária ('Yes')

# O modelo aprende com os dados de treino
model.fit(X_train, y_train)
print("Modelo treinado com sucesso!")

# Previsão e Avaliação da Performance

# O modelo faz previsões nos dados de teste, que ele nunca viu antes
y_pred = model.predict(X_test)

# Comparando as previsões (y_pred) com os resultados reais (y_test)
print("\n--- Resultados da Avaliação do Modelo ---")
print(f"Acurácia: {accuracy_score(y_test, y_pred):.2f}")

print("\nMatriz de Confusão:")
print(confusion_matrix(y_test, y_pred))

print("\nRelatório de Classificação:")
print(classification_report(y_test, y_pred))
```

```python

# 5. ANÁLISE DOS RESULTADOS E RECOMENDAÇÕES DE NEGÓCIO

# Extraindo a Importância das Features do Modelo

# O modelo treinado armazena a importância de cada feature.
importances = model.feature_importances_
feature_names = X.columns

# DataFrame do Pandas para facilitar a visualização
feature_importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})

# Ordenando o DataFrame pela importância em ordem decrescente
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# Selecionando as 15 features mais importantes para o gráfico
top_10_features = feature_importance_df.head(10)

print("As 10 features mais importantes segundo o modelo:")
display(top_10_features)

# Visualizando a Importância das Features
plt.figure(figsize=(12, 8))
sns.barplot(x='Importance', y='Feature', data=top_10_features, palette='Oranges')

plt.title('Top 15 Fatores que Influenciam a Rotatividade', fontsize=16)
plt.xlabel('Importância Relativa', fontsize=12)
plt.ylabel('Fator', fontsize=12)
plt.show()

# Salvando os dados de importância em um arquivo CSV.
feature_importance_df.to_csv('../data/feature_importances.csv', index=False)

print("Arquivo com a importância das features salvo com sucesso em 'data/feature_importances.csv'")
```
